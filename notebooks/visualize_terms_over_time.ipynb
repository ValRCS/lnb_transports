```python
# let's see how many times these lemma are presently in term column
for lemma in tqdm(lemmas):
    # count the number of times the lemma is used
    count = df[df['term'] == lemma].shape[0]
    print(f"Number of times {lemma} is used: {count}")
```

```python
# let's use first lemma for all other lemmas in term column
# we will use the first lemma in the list
first_lemma = lemmas[0]
# let's go the other lemmas one by one and replace them with the first lemma
for lemma in tqdm(lemmas[1:]):
    # replace the lemma with the first lemma
    df.loc[df['term'] == lemma, 'term'] = first_lemma
# let's see how many times these lemma are now in term column
for lemma in tqdm(lemmas):
    # count the number of times the lemma is used
    count = df[df['term'] == first_lemma].shape[0]
    print(f"Number of times {first_lemma} is used: {count}")
```

### Ploting legend for size

We need to plot a separate legend for the above graph

Our minimal size is 6 representing 5 + 1 for the smallest size when we have 1 occurence
Our maximal size would be then 5 + max(occurences) for the largest size when we have the most occurences

```python
def get_top_n_authors_df(df, term, n=5):
    # find top n authors with most term
    top_authors = df[df.lemma == term].groupby('author').size().nlargest(n)
    # create a histogram of top n authors with most term
    vis_df = top_authors.head(n)
    return vis_df
# let's get trace given we already have a authors_df
def get_authors_trace(authors_df, term, y_range_max=None, font_size=18, color="rgb(0, 61, 165)", term_dict=None):
    if term_dict:
        name = term_dict[term]
        print(f"Will use name: {name} instead of {term}")
    else:
        name = term
        print(f"Will use name: {name} as is {term}")
    
    trace = go.Bar(
        x=authors_df.index,
        y=authors_df.values,
        name=name,
        # font size
        textfont=dict(size=font_size),
        # marker color
        marker_color=color,
    )
    return trace
# let's create a function that takes df and term_list and returns a figure with subplots for each term
def term_top_authors_subplots(df, 
                              term_list, 
                              top_n = 5, 
                              sort=True, 
                              height_per_trace=300, 
                              columns=3, 
                              get_y_range_max=True, 
                              turn_off_legend=True,
                              font_size=18,
                              width = 1242,
                              term_dict=None,
                              top_authors_dfs=None):
    if not top_authors_dfs:
        top_authors_dfs = [get_top_n_authors_df(df, term, top_n) for term in term_list]
    # create a list of traces
    # get top_n authors for each term as df
    if sort:
        # we will sort term_list alphabetically
        # however we also need to match top_authors_dfs to term_list
        # so we will sort top_authors_dfs as well
        term_list, top_authors_dfs = zip(*sorted(zip(term_list, top_authors_dfs)))

    if get_y_range_max:
        # get maximum y value
        y_max = max([df.max() for df in top_authors_dfs])
        # round up to nearest 10
        y_max = math.ceil(y_max / 10) * 10
    else:
        y_max = None
    traces = []

    for term, authors_df in tqdm(zip(term_list, top_authors_dfs)):
        trace = get_authors_trace(authors_df, term, y_max, term_dict=term_dict)
        traces.append(trace)
    # create a subplot
    fig = make_subplots(rows=math.ceil(len(term_list)/columns), cols=columns, subplot_titles=term_list)
    fig.update_layout(height=height_per_trace * len(term_list) // columns)
    # width
    fig.update_layout(width=width)
    if y_max:
        # fig.update_layout(yaxis=dict(range=[0, y_max]))
        print(f"update yaxis range to [0, {y_max}]")
        fig.update_yaxes(range=[0, y_max])
    # fig.print_grid()
    # add each fig to the subplot
    for i, trace in enumerate(traces):
        fig.append_trace(trace, row=i//columns+1, col=i%columns+1)
    # update height to 3000
    # fig.update_layout(height=3000)
    # return fig
    # change font size
    fig.update_layout(font=dict(size=font_size, color="black"))
    if turn_off_legend:
        fig.update_layout(showlegend=False)
    # update annotation font size
    fig.update_layout(annotations=[dict(font=dict(size=font_size))])
    return fig
```

```python
# let's create a function that creates a pie chart given a percentage
# the result should be a plotly go pie chart partially filled with color

def create_pie_chart(percentage, color="gray", hole=0.0, annotation="2024", annotation_font_size=24):
    # create a pie chart
    fig = go.Figure()
    fig.add_trace(go.Pie(
        labels=[f"{percentage:.2f}%", f"{100-percentage:.2f}%"],
        values=[percentage, 100-percentage],
        hole=hole,
        marker_colors=[color, "white"],
        textinfo="none"
    ))
    # turn off labels
    fig.update_layout(showlegend=False)
    # add annotation below the pie chart
    fig.add_annotation(
        x=0.5,
        y=0,
        text=annotation,
        showarrow=False,
        font=dict(size=annotation_font_size, color="black"),
        yshift=-20,
        # we want black color for annotation


    )
    return fig

# let's test the function
# fig = create_pie_chart(50, annotation_font_size=48)
# fig.show()
# now let's create a function that creates a pie chart for each year given a list of percentages and  years
# we want to use subplots to create a separate pie chart for each year
# we will use plotly graph objects to create the same chart

def create_pie_charts(percentages, 
                      years, 
                      pies_in_row=5, 
                      color="black", 
                      hole=0.0, 
                      height_per_row=200, 
                      width=800,
                      annotation_font_size=24):
    # create a figure
    rows = len(percentages)//pies_in_row+1
    cols = pies_in_row
    fig = make_subplots(rows=rows, 
                        cols=cols,
                        subplot_titles=years,
                        specs=[[{'type':'domain'}]*cols]*rows,
    )
    # create a pie chart for each year
    for i, (percentage, year) in enumerate(zip(percentages, years)):
        # create a pie chart
        pie = create_pie_chart(percentage, color, hole, year, annotation_font_size=annotation_font_size)
        # add pie chart to the subplot
        fig.add_trace(pie.data[0], row=(i)//pies_in_row+1, col=(i)%pies_in_row+1)

    # use annotation font size
    fig.update_layout(annotations=[dict(font=dict(size=annotation_font_size))])
    # hide legend
    fig.update_layout(showlegend=False)
    # set width
    fig.update_layout(width=width)
    # set height to height_per_row * rows
    fig.update_layout(height=height_per_row * rows)
    return fig

# let's test the function
# percentages = [40, 50, 60, 70, 55, 67, 35]
# years = [2020, 2021, 2022, 2023, 2024,2025,2026]
# fig = create_pie_charts(percentages, years, annotation_font_size=32, height_per_row=200)
# fig.show()
```
